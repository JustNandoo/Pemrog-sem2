```python
class DataRecord:
    tax_rate = 0.10

    def __init__(self, product_name, quantity, price):
        self.product_name = product_name
        self.quantity = quantity
        self.price = price

    def total_price(self):
        return self.quantity * self.price

    def total_price_tax(self):
        total_price = self.total_price()
        total_with_tax = total_price + (total_price * DataRecord.tax_rate)
        return total_with_tax

    def __str__(self):
        return (
            f"Name: {self.product_name}; "
            f"Qty: {self.quantity}; "
            f"Price: {self.price:,.0f}"
        )


data_record_1 = DataRecord("Buku Tulis", 11, 25500)
data_record_2 = DataRecord("Sticky Notes", 5, 8000)
data_record_3 = DataRecord("Rautan Pensil", 3, 4000)

print(f"Total sales (without tax) for {data_record_1.product_name}: Rp. {data_record_1.total_price():,.0f}")
print(f"Total sales (without tax) for {data_record_2.product_name}: Rp. {data_record_2.total_price():,.0f}")
print(f"Total sales (with tax, e.g.: 10%) for {data_record_3.product_name}: Rp. {data_record_3.total_price_tax():,.0f}")

print(data_record_1)
print(data_record_2)
print(data_record_3)
```

```python
import re

class DataRecord:
    def __init__(self, product_name, quantity, price):
        self.product_name = product_name
        self.quantity = quantity
        self.price = price


class Dataset:
    def __init__(self):

        self.records = []
        self.name = "Dataset Penjualan"

    def load_data_from_file(self, filename):

        try:
            with open(filename, "r") as file:
                total_data = 0

                for line in file:
                    line = line.strip()

                    if line == "":
                        continue

                    pattern = r"^(.+),\s*(\d+),\s*(\d+)$"
                    match = re.match(pattern, line)

                    if match:
                        product_name = match.group(1)
                        quantity = int(match.group(2))
                        price = int(match.group(3))

                        record = DataRecord(product_name, quantity, price)
                        self.add_data(record)
                        total_data += 1

                return total_data

        except FileNotFoundError:
            print("File tidak ditemukan")
            return None

    def add_data(self, record):
        if isinstance(record, DataRecord):
            self.records.append(record)
            return True

        return False

    def display_data(self):

        if self.records:
            for record in self.records:
                print(
                    f"Name: {record.product_name}; "
                    f"Qty: {record.quantity}; "
                    f"Price: {record.price:,}"
                )
        else:
            print("Tidak ada data")


dataset = Dataset()

data1 = DataRecord("Buku tulis", 11, 25500)
data2 = DataRecord("Pulpen", 20, 3500)

dataset.add_data(data1)
dataset.add_data(data2)

dataset.display_data()
```

```python
import re

class DataRecord:
    def __init__(self, product_name, quantity, price):
        self.product_name = product_name
        self.quantity = quantity
        self.price = price


class Dataset:
    def __init__(self):
        self.records = []
        self.name = "Dataset Penjualan"

    def load_data_from_file(self, filename):
        try:
            total_data = 0

            with open(filename, "r") as file:
                for line in file:
                    line = line.strip()

                    if line == "":
                        continue

                    pattern = r"^(.+)\s\((\d+)\)\s-\s(\d+)$"
                    match = re.match(pattern, line)

                    if match:
                        product_name = match.group(1)
                        quantity = int(match.group(2))
                        price = int(match.group(3))

                        record = DataRecord(product_name, quantity, price)
                        self.records.append(record)
                        total_data += 1

            print(f"{total_data} from {filename} added to dataset_beta")
            return total_data

        except FileNotFoundError:
            print(f"File {filename} is not found")
            return None

    def add_data(self, record):
        if isinstance(record, DataRecord):
            self.records.append(record)
            print(f"{record.product_name} added to dataset_alpha")
            return True
        else:
            print(f"{record} is not object of DataRecord")
            return False

    def display_data(self):
        if len(self.records) == 0:
            print("Tidak ada data")
        else:
            for i, record in enumerate(self.records, start=1):
                print(
                    f"{i}. Name: {record.product_name}; "
                    f"Qty: {record.quantity}; "
                    f"Price: {record.price:,}"
                )


data_record_1 = DataRecord("Buku tulis", 11, 25500)
data_record_2 = DataRecord("Sticky notes", 5, 8000)
data_record_3 = DataRecord("Rautan pensil", 3, 4000)

dataset_alpha = Dataset()
dataset_beta = Dataset()

dataset_alpha.add_data(data_record_1)
dataset_alpha.add_data(data_record_1)
dataset_alpha.add_data(data_record_2)
dataset_alpha.add_data(data_record_3)

dataset_alpha.add_data("Correction Tape")

dataset_beta.load_data_from_file("../data/study/data_penjualan.txt")

print("\ndataset_alpha:")
dataset_alpha.display_data()

print("\ndataset_beta:")
dataset_beta.display_data()
```

```python
class DataAnalyzer:
    def __init__(self, dataset):
        self.dataset = dataset

    def calculate_total_sales(self, include_tax=False):
        if not self.dataset.records:
            return None

        total_sales = sum(
            record.quantity * record.price
            for record in self.dataset.records
        )

        if include_tax:
            total_sales *= 1.10

        return total_sales

    def find_best_selling(self):
        if not self.dataset.records:
            return None

        best_selling = max(
            self.dataset.records,
            key=lambda record: record.quantity
        )

        return best_selling.product_name

    def calculate_average_price(self):
        if not self.dataset.records:
            return None

        total_price = sum(
            record.price
            for record in self.dataset.records
        )

        return total_price / len(self.dataset.records)

analyzer_alpha = DataAnalyzer(dataset_alpha)
analyzer_beta = DataAnalyzer(dataset_beta)

print("Analysis of dataset_alpha")
print(f"Total sales (without tax): {analyzer_alpha.calculate_total_sales():,.0f}")
print(f"Best selling item: {analyzer_alpha.find_best_selling()}")
print(f"Average price: {analyzer_alpha.calculate_average_price():,.0f}")

print("Analysis of dataset_beta")
print(f"Total sales (with tax): {analyzer_beta.calculate_total_sales(include_tax=True):,.0f}")
print(f"Best selling item: {analyzer_beta.find_best_selling()}")
print(f"Average price: {analyzer_beta.calculate_average_price():,.0f}")
```
