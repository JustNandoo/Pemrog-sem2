````python
# Simulasi membuat modul sendiri dengan menulis file Python
modul_code = '''
"""Modul utilitas untuk perhitungan nilai mahasiswa."""

def hitung_rata_rata(nilai_list):
    """Menghitung rata-rata dari list nilai."""
    if not nilai_list:
        return 0
    return sum(nilai_list) / len(nilai_list)

def konversi_huruf(nilai):
    """Mengkonversi nilai angka ke huruf."""
    if nilai >= 85:  return "A"
    elif nilai >= 75: return "B"
    elif nilai >= 65: return "C"
    elif nilai >= 55: return "D"
    else:             return "E"

PASSING_GRADE = 65
'''

# Tulis modul ke file
with open("utilitas_nilai.py", "w") as f:
    f.write(modul_code)
print("Modul 'utilitas_nilai.py' berhasil dibuat!")
````

# 🔍 MATERI 3: REGULAR EXPRESSION (REGEX)

Regex adalah pola (*pattern*) yang digunakan untuk mencocokkan teks.

| Pola | Fungsi |
|------|--------|
| `.` | Karakter apa pun (kecuali `\n`) |
| `\d` | Digit angka `[0-9]` |
| `\w` | Karakter kata `[a-zA-Z0-9_]` |
| `\s` | Spasi/whitespace |
| `\D`, `\W`, `\S` | Kebalikannya |
| `^` | Awal string |
| `$` | Akhir string |
| `[]` | Kumpulan karakter |
| `+` | 1 atau lebih |
| `*` | 0 atau lebih |
| `?` | 0 atau 1 |
| `{n}` | Tepat n kali |
| `{n,m}` | n sampai m kali |
| `()` | Pengelompokan |

---

## 3.1 Fungsi-Fungsi Dasar Regex

````python
import re

# search() : cari pola di mana saja dalam string
text = "NRP mahasiswa adalah 3123600001 dan 3123600002"

hasil = re.search(r'\d', text)  # cari digit pertama
print("=== search() ===")
print(f"Match object: {hasil}")
print(f"Span: {hasil.span()}")       # (start, end)
print(f"Teks match: {hasil.group()}") # digit pertama yang ditemukan

# Perhatikan: search() berhenti di kecocokan PERTAMA
print(f"\nMeskipun ada banyak digit, search() hanya menemukan: '{hasil.group()}'")
````

````python
# match() : cari pola hanya di AWAL string
pola = r'^\d'
teks1 = "31236abc"
teks2 = "abc31236"

print("=== match() ===")
print(f"'{teks1}' dimulai angka? {re.match(pola, teks1) is not None}")
print(f"'{teks2}' dimulai angka? {re.match(pola, teks2) is not None}")
````

````python
# fullmatch() : pola harus cocok SELURUH string
pola_nrp = r'3\d{9}'  # dimulai 3, diikuti tepat 9 digit

test_nrp = ["3123600001", "3123600", "31236000010", "1234567890"]

print("=== fullmatch() untuk validasi NRP ===")
for nrp in test_nrp:
    cocok = re.fullmatch(pola_nrp, nrp)
    status = "✅ Valid" if cocok else "❌ Tidak valid"
    print(f"  '{nrp}': {status}")
````

````python
# findall() : temukan SEMUA kecocokan, kembalikan sebagai list
text = "NRP: 3123600001, 3123600002, dan 3123600003 terdaftar"

semua_nrp = re.findall(r'\d{10}', text)
print("=== findall() ===")
print(f"Semua NRP ditemukan: {semua_nrp}")

# finditer() : seperti findall tapi kembalikan iterator objek match
print("\n=== finditer() ===")
for m in re.finditer(r'\d{10}', text):
    print(f"  Found: '{m.group()}' at span {m.span()}")
````

````python
# sub() : ganti kecocokan dengan string lain (seperti find & replace)
text = "Nilai Ahmad: 85, Nilai Budi: 78, Nilai Citra: 92"

# Sembunyikan semua angka
hasil_sensor = re.sub(r'\d+', '***', text)
print("=== sub() ===")
print(f"Original : {text}")
print(f"Tersensor: {hasil_sensor}")

# split() : pecah string berdasarkan pola
teks_csv = "Ahmad,85;Budi 78|Citra:92"
hasil_split = re.split(r'[,;|: ]+', teks_csv)
print(f"\n=== split() ===")
print(f"Input : {teks_csv}")
print(f"Output: {hasil_split}")
````

## 3.2 Pola Regex - Penjelasan & Contoh

````python
import re

# --- ^ dan $ (awal dan akhir string) ---
nama = ["apple", "ap", "APPLE", "Tap", "pineapple"]

print("=== Pola '^ap' (dimulai dengan 'ap') ===")
for n in nama:
    cocok = re.search(r'^ap', n)
    print(f"  '{n}': {'✅' if cocok else '❌'}")

# Case insensitive dengan re.IGNORECASE
print("\n=== Pola '^ap' dengan IGNORECASE ===")
for n in nama:
    cocok = re.search(r'^ap', n, re.IGNORECASE)
    print(f"  '{n}': {'✅' if cocok else '❌'}")
````

````python
# --- . (titik) sebagai wildcard ---
nama2 = ["Diana", "dian", "anabel", "Fanianto", "banana"]

print("=== Pola '.an.' (ada karakter-'an'-karakter) ===")
for n in nama2:
    cocok = re.search(r'.an.', n)
    print(f"  '{n}': {'✅' if cocok else '❌'}")

print("\n=== Pola '.an$' (diakhiri 'an') ===")
for n in nama2:
    cocok = re.search(r'.an$', n)
    print(f"  '{n}': {'✅' if cocok else '❌'}")
````

````python
# --- \d dan pengulangan ---
teks_list = ["ega31", "kian11", "sesa4", "daniel7", "lely99", "nama"]

print("=== Pola '\\d\\d$' (diakhiri 2 digit) ===")
for t in teks_list:
    cocok = re.search(r'\d\d$', t)
    print(f"  '{t}': {'✅' if cocok else '❌'}")

print("\n=== Pola '^@\\w' (dimulai @ lalu karakter kata) ===")
username = ["@ahmad", "sesa4", "@ kian11", "@_Gita", "@123"]
for u in username:
    cocok = re.search(r'^@\w', u)
    print(f"  '{u}': {'✅' if cocok else '❌'}")
````

````python
# --- Kurung siku [] ---
print("=== [aeiou] - karakter vokal ===")
kata = ["ahmad", "budi", "xyz", "aeiou"]
for k in kata:
    cocok = re.search(r'[aeiou]', k)
    print(f"  '{k}' mengandung vokal: {'✅' if cocok else '❌'}")

# [^...] = negasi
print("\n=== [^\\d_] tidak dimulai digit atau underscore ===")
identifier = ["nama", "123abc", "_private", "PublicVar", "3data"]
for i in identifier:
    cocok = re.search(r'^[^\d_]', i)
    print(f"  '{i}': {'✅' if cocok else '❌'}")
````

````python
# --- Pengulangan +, *, ?, {n}, {n,m} ---
print("=== Pola '\\d+' (1 atau lebih digit) ===")
teks = ["abc", "a1b", "123", ""]
for t in teks:
    cocok = re.search(r'\d+', t)
    print(f"  '{t}': {cocok.group() if cocok else 'Tidak ditemukan'}")

# Validasi nomor telepon Indonesia: 08 diikuti 8-11 digit
print("\n=== Validasi Nomor HP (^08\\d{8,11}$) ===")
nomor_hp = ["081234567890", "0812345", "08123456789012", "081-234-5678", "08222222222"]
for hp in nomor_hp:
    cocok = re.fullmatch(r'^08\d{8,10}$', hp)
    print(f"  '{hp}': {'✅ Valid' if cocok else '❌ Tidak valid'}")
````

````python
# --- Kurung () untuk pengelompokan ---
# Pola email institusi: <huruf+>.<pens|ppns|its>.ac.id
pola_email_institusi = r'^[a-zA-Z][\w.]+@[\w]+\.(pens|ppns|its)\.ac\.id$'

email_list = [
    "ahmad.fauzi@student.pens.ac.id",
    "budi.s@student.ppns.ac.id",
    "citra@student.its.ac.id",
    "fajar.n@gmail.com",
    "kian11@yahoo.com",
    "99lely@student.pens.ac.id",   # tidak valid: dimulai angka
]

print("=== Validasi Email Institusi (pens/ppns/its) ===")
for email in email_list:
    cocok = re.fullmatch(pola_email_institusi, email)
    print(f"  {email}: {'✅ Institusi' if cocok else '❌ Non-institusi/Invalid'}")
````

````python
# --- Menghindari karakter khusus dengan \ ---
print("=== Escape karakter khusus ===")
teks_harga = ["Rp.15000", "Rp.75000", "USD100", ".500gr", "25000"]

# Tanpa escape: '^.' cocok semua (. = karakter apapun)
print("Pola '^.' (tanpa escape - titik = apapun):")
for t in teks_harga:
    cocok = re.search(r'^.', t)
    print(f"  '{t}': {'✅' if cocok else '❌'}")

# Dengan escape: '\.' cocok hanya titik literal
print("\nPola '^\\.' (dengan escape - titik literal):")
for t in teks_harga:
    cocok = re.search(r'^\.', t)
    print(f"  '{t}': {'✅' if cocok else '❌'}")
````

## 3.4 Regex dari File CSV

````python
import pandas as pd
import re

df = pd.read_csv("../data/hasil_generate/data_mahasiswa.csv")

# 1. Filter email institusi menggunakan regex di pandas
pola_institusi = r'[\w.]+@[\w]+\.(pens|ppns|its)\.ac\.id'

df["email_institusi"] = df["email"].apply(
    lambda e: bool(re.fullmatch(pola_institusi, str(e)))
)

print("=== Klasifikasi Email ===")
print(df[["nama", "email", "email_institusi"]].to_string(index=False))
````

````python
# 2. Validasi nomor telepon
pola_hp = r'^(08\d{8,10}|031-\d{7})$'

df["telepon_valid"] = df["telepon"].apply(
    lambda t: bool(re.fullmatch(pola_hp, str(t)))
)

print("=== Validasi Nomor Telepon ===")
print(df[["nama", "telepon", "telepon_valid"]].to_string(index=False))
print(f"\nValid: {df['telepon_valid'].sum()}, Tidak valid: {(~df['telepon_valid']).sum()}")
````

````python
# 3. Ekstrak domain email
def ekstrak_domain(email):
    match = re.search(r'@([\w.]+)', str(email))
    return match.group(1) if match else "tidak_diketahui"

df["domain_email"] = df["email"].apply(ekstrak_domain)

print("=== Distribusi Domain Email ===")
print(df["domain_email"].value_counts().to_string())
````

````python
# ============================================================
# STUDI KASUS: Sistem Pengolahan Data Mahasiswa
# Menggunakan: Library (pandas, math, os) + File Handling + Regex
# ============================================================
import pandas as pd
import math
import re
import os
from datetime import datetime

# ---- STEP 1: Baca data dari CSV (FILE HANDLING + LIBRARY) ----
print("=" * 60)
print("STEP 1: Membaca Data")
print("=" * 60)
df = pd.read_csv("data_mahasiswa.csv")
print(f"✅ Berhasil membaca {len(df)} data mahasiswa")
print(df.head(3).to_string(index=False))
````

````python
# ---- STEP 2: Validasi data dengan REGEX ----
print("\n" + "=" * 60)
print("STEP 2: Validasi Data dengan Regex")
print("=" * 60)

# Pola validasi
POLA_NRP      = r'^3\d{9}$'
POLA_EMAIL    = r'^[a-zA-Z][\w.]+@[\w]+\.[\w.]+$'
POLA_TELEPON  = r'^(08\d{8,10}|031-\d{7})$'

# Fungsi validasi
def validasi(nilai, pola):
    return bool(re.fullmatch(pola, str(nilai)))

df["nrp_valid"]     = df["nrp"].apply(lambda x: validasi(x, POLA_NRP))
df["email_valid"]   = df["email"].apply(lambda x: validasi(x, POLA_EMAIL))
df["telepon_valid"] = df["telepon"].apply(lambda x: validasi(x, POLA_TELEPON))
df["data_lengkap"]  = df["nrp_valid"] & df["email_valid"] & df["telepon_valid"]

print("Status Validasi:")
print(f"  NRP valid     : {df['nrp_valid'].sum()}/{len(df)}")
print(f"  Email valid   : {df['email_valid'].sum()}/{len(df)}")
print(f"  Telepon valid : {df['telepon_valid'].sum()}/{len(df)}")
print(f"  Data lengkap  : {df['data_lengkap'].sum()}/{len(df)}")

masalah = df[~df["data_lengkap"]][["nama", "nrp_valid", "email_valid", "telepon_valid"]]
print("\nData bermasalah:")
print(masalah.to_string(index=False))
````

````python
# ---- STEP 3: Hitung nilai akhir menggunakan LIBRARY (math, pandas) ----
print("\n" + "=" * 60)
print("STEP 3: Kalkulasi Nilai")
print("=" * 60)

# Hitung nilai akhir (bobot: UTS 30%, UAS 40%, Tugas 30%)
df["nilai_akhir"] = (
    df["nilai_uts"] * 0.30 +
    df["nilai_uas"] * 0.40 +
    df["nilai_tugas"] * 0.30
).round(2)

# Grade dengan modul utilitas yang dibuat sebelumnya
import utilitas_nilai  # modul yang kita buat di bagian Library
df["grade"] = df["nilai_akhir"].apply(utilitas_nilai.konversi_huruf)
df["lulus"] = df["nilai_akhir"] >= utilitas_nilai.PASSING_GRADE

# Statistik menggunakan math
nilai_all = df["nilai_akhir"].tolist()
rata_rata  = sum(nilai_all) / len(nilai_all)
variansi   = sum((x - rata_rata)**2 for x in nilai_all) / len(nilai_all)
std_dev    = math.sqrt(variansi)

print(f"Rata-rata nilai akhir : {rata_rata:.2f}")
print(f"Standar deviasi       : {std_dev:.2f}")
print(f"Nilai tertinggi       : {max(nilai_all)}")
print(f"Nilai terendah        : {min(nilai_all)}")
print(f"Jumlah lulus          : {df['lulus'].sum()}/{len(df)}")
print("\nDistribusi Grade:")
print(df["grade"].value_counts().sort_index().to_string())
````

````python
# ---- STEP 4: Baca data tambahan dari TXT dan gabungkan ----
print("\n" + "=" * 60)
print("STEP 4: Ekstraksi Data dari TXT (FILE HANDLING + REGEX)")
print("=" * 60)

# Baca file TXT
with open("data_mahasiswa.txt", "r") as f:
    baris_txt = f.readlines()

# Ekstrak data dari TXT menggunakan regex
data_dari_txt = []
for baris in baris_txt:
    # Pola: Nama: X | NRP: Y | Email: Z | Nilai: W
    match = re.search(
        r'Nama: ([\w ]+) \| NRP: (\d+) \| Email: ([\w.@]+) \| Nilai: (\d+)',
        baris
    )
    if match:
        data_dari_txt.append({
            "nama_txt"   : match.group(1).strip(),
            "nrp_txt"    : match.group(2),
            "email_txt"  : match.group(3),
            "nilai_txt"  : int(match.group(4))
        })

df_txt = pd.DataFrame(data_dari_txt)
print(f"✅ Berhasil ekstrak {len(df_txt)} data dari TXT")
print(df_txt.head(3).to_string(index=False))
````

````python
# ---- STEP 5: Simpan laporan ke file TXT dan CSV (FILE HANDLING) ----
print("\n" + "=" * 60)
print("STEP 5: Simpan Laporan")
print("=" * 60)

# Simpan laporan ringkasan ke TXT
waktu = datetime.now().strftime("%d-%m-%Y %H:%M:%S")
lulus_count    = df["lulus"].sum()
tidak_lulus    = len(df) - lulus_count

laporan_txt = f"""LAPORAN NILAI MAHASISWA PEMROGRAMAN 2
PENS - 2025
Dibuat: {waktu}
{'='*45}

STATISTIK UMUM
  Total Mahasiswa  : {len(df)}
  Rata-rata Nilai  : {rata_rata:.2f}
  Standar Deviasi  : {std_dev:.2f}
  Nilai Tertinggi  : {max(nilai_all)}
  Nilai Terendah   : {min(nilai_all)}
  Lulus            : {lulus_count}
  Tidak Lulus      : {tidak_lulus}

DISTRIBUSI GRADE
"""

for grade, count in df["grade"].value_counts().sort_index().items():
    laporan_txt += f"  Grade {grade}: {count} mahasiswa\n"

laporan_txt += f"\nDATA VALIDASI\n"
laporan_txt += f"  Data bermasalah: {(~df['data_lengkap']).sum()} mahasiswa\n"
for _, row in df[~df["data_lengkap"]].iterrows():
    laporan_txt += f"  - {row['nama']}: "
    masalah_detail = []
    if not row["nrp_valid"]:     masalah_detail.append("NRP tidak valid")
    if not row["email_valid"]:   masalah_detail.append("Email tidak valid")
    if not row["telepon_valid"]: masalah_detail.append("Telepon tidak valid")
    laporan_txt += ", ".join(masalah_detail) + "\n"

# Tulis laporan TXT
with open("laporan_nilai.txt", "w") as f:
    f.write(laporan_txt)
print("✅ Laporan TXT disimpan: laporan_nilai.txt")
print()
print(laporan_txt)
````

````python
# Simpan data lengkap ke CSV
kolom_output = ["id", "nama", "nrp", "jurusan", "email", "telepon",
                "nilai_uts", "nilai_uas", "nilai_tugas", "nilai_akhir",
                "grade", "lulus", "email_valid", "telepon_valid"]

df[kolom_output].to_csv("laporan_lengkap.csv", index=False)
print("✅ Laporan CSV disimpan: laporan_lengkap.csv")

# Simpan juga data yang bermasalah secara terpisah
df[~df["data_lengkap"]][kolom_output].to_csv("data_bermasalah.csv", index=False)
print("✅ Data bermasalah disimpan: data_bermasalah.csv")

# Verifikasi file yang dibuat
file_output = ["laporan_nilai.txt", "laporan_lengkap.csv", "data_bermasalah.csv"]
print("\nFile yang dihasilkan:")
for fname in file_output:
    size = os.path.getsize(fname)
    print(f"  {fname}: {size} bytes")
````

````python
# ---- TAMPILAN AKHIR: Rekap lengkap ----
print("\n" + "=" * 60)
print("REKAP AKHIR SEMUA MAHASISWA")
print("=" * 60)
tampil = df[["nama", "jurusan", "nilai_akhir", "grade", "lulus", "data_lengkap"]].copy()
tampil["lulus"]       = tampil["lulus"].map({True: "✅", False: "❌"})
tampil["data_lengkap"] = tampil["data_lengkap"].map({True: "✅", False: "⚠️"})
tampil.columns = ["Nama", "Jurusan", "Nilai Akhir", "Grade", "Lulus", "Data OK"]
print(tampil.to_string(index=False))
````
