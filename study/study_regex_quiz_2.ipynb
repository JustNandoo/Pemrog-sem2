```python
import re

with open("../data/praktikum/berita.txt", "r") as file:
    text = file.readline()

pattern = r'\b(?:[A-Z]{2,}|(?:[A-Z]\.))\b'


singkatan = re.findall(pattern, text)

list_singkatan = list(set(singkatan))
print(list_singkatan)


```

```python
import re

with open("../data/praktikum/berita.txt", "r") as file:
    text = file.read()

pattern = r'(?<=\b[Jj]udul [Oo]rasi [Ii]lmiah )[A-Z][^.]+(?=\.)'

orasi = re.findall(pattern, text)

list_orasi = list(set(orasi))
print(list_orasi)
```

```python
import re

with open("../data/praktikum/berita.txt", "r") as file:
    text = file.read()

pattern = r'Prof\..+?([A-Z][a-z]+(?:\s[A-Z][a-z]+)+)'

names = re.findall(pattern, text)

sorted_names = sorted(set(names))

print(sorted_names)
```

```python
import pandas as pd
import re

df = pd.read_csv("../data/praktikum/reviews_dataset_updated.csv")

def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

df['punctuation_and_digit_removal'] = df['review_text'].apply(clean_text)

df['casefolding'] = df['punctuation_and_digit_removal'].str.lower()

df.to_csv("cleaned.csv", index=False)

print(df.head())
```
