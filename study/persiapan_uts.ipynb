## Study Case 1 - Validasi Nama dengan Regex (data langsung di kode)
Memeriksa apakah nama diawali huruf dan hanya berisi huruf/spasi.


```python
pattern_nama = r'^[A-Za-z][A-Za-z ]+$'

daftar_nama = [
    "Andi",
    "Budi Santoso",
    "Citra Lestari",
    "D1na",
    "Eko_Nugroho",
    " fajar"
]

for nama in daftar_nama:
    if re.match(pattern_nama, nama):
        print(f"{nama} -> valid")
    else:
        print(f"{nama} -> tidak valid")
```

## Study Case 2 - Validasi Email Domain Tertentu (data langsung di kode)
Email harus sesuai pola `@pemrograman.id`.


```python
pattern_email = r'^[A-Za-z][A-Za-z0-9_]{1,}@pemrograman\.id$'

daftar_email = [
    "andi_01@pemrograman.id",
    "budi_dev@pemrograman.id",
    "citra123@gmail.com",
    "1dina@pemrograman.id",
    "eko-program@pemrograman.id"
]

for email in daftar_email:
    if re.match(pattern_email, email):
        print(email, "-> valid")
    else:
        print(email, "-> tidak valid")
```

## Study Case 3 - Validasi Data Student (data langsung di kode)
Mengecek nama, NIM, kelas, dan email kampus.


```python
pattern_nim = r'^2024\d{3}$'
pattern_kelas = r'^[A-Z]{2}-\d[A-Z]$'
pattern_email_kampus = r'^[a-z]+@student\.kampus\.ac\.id$'

students = [
    {"nama": "andi", "nim": "2024001", "kelas": "TI-1A", "email": "andi@student.kampus.ac.id"},
    {"nama": "budi", "nim": "24A002", "kelas": "TI-1B", "email": "budi@student.kampus.ac.id"},
    {"nama": "citra", "nim": "2024003", "kelas": "SI-2A", "email": "citra@student.kampus.ac.id"},
    {"nama": "dina", "nim": "2024004", "kelas": "si-2a", "email": "dina@student.kampus.ac.id"},
]

for s in students:
    hasil_nim = "valid" if re.match(pattern_nim, s["nim"]) else "tidak valid"
    hasil_kelas = "valid" if re.match(pattern_kelas, s["kelas"]) else "tidak valid"
    hasil_email = "valid" if re.match(pattern_email_kampus, s["email"]) else "tidak valid"
    print(s["nama"], "| NIM:", hasil_nim, "| Kelas:", hasil_kelas, "| Email:", hasil_email)
```

## Study Case 4 - Try Except Input Angka

```python
try:
    a = float(input("Masukkan angka pertama: "))
    b = float(input("Masukkan angka kedua: "))
    hasil = a / b
    print("Hasil pembagian =", hasil)
except ValueError:
    print("Error: input harus berupa angka")
except ZeroDivisionError:
    print("Error: tidak bisa dibagi dengan nol")
except Exception as e:
    print("Terjadi error:", e)
```

## Study Case 5 - Membaca File Teks lalu Ekstrak Email dengan Regex
File yang digunakan: `emails_teks.txt`.


```python
pattern_email_umum = r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}'

try:
    with open("emails_teks.txt", "r", encoding="utf-8") as f:
        isi = f.read()
    hasil = re.findall(pattern_email_umum, isi)
    print("Email yang ditemukan:")
    for email in hasil:
        print("-", email)
except FileNotFoundError:
    print("File emails_teks.txt tidak ditemukan")
except Exception as e:
    print("Terjadi error:", e)
```

## Study Case 6 - Membaca CSV Pendaftaran dan Validasi Regex
File yang digunakan: `pendaftaran_dummy.csv`.


```python
pattern_nama = r'^[A-Za-z][A-Za-z ]+$'
pattern_email = r'^[A-Za-z][A-Za-z0-9_]{1,}@pemrograman\.id$'

try:
    with open("pendaftaran_dummy.csv", "r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            nama_valid = bool(re.match(pattern_nama, row["nama"]))
            email_valid = bool(re.match(pattern_email, row["email"]))
            print(row["nama"], "| nama:", nama_valid, "| email:", email_valid)
except FileNotFoundError:
    print("File pendaftaran_dummy.csv tidak ditemukan")
except Exception as e:
    print("Terjadi error:", e)
```

## Study Case 7 - Membaca CSV Student lalu Filter Data Valid
File yang digunakan: `students_dummy.csv`.


```python
pattern_nim = r'^2024\d{3}$'
pattern_kelas = r'^[A-Z]{2}-\d[A-Z]$'
pattern_email = r'^[a-z]+@student\.kampus\.ac\.id$'

data_valid = []

try:
    with open("students_dummy.csv", "r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            if (re.match(pattern_nim, row["nim"]) and
                re.match(pattern_kelas, row["kelas"]) and
                re.match(pattern_email, row["email"])):
                data_valid.append(row)

    print("Student yang valid:")
    for s in data_valid:
        print(s)
except FileNotFoundError:
    print("File students_dummy.csv tidak ditemukan")
except Exception as e:
    print("Terjadi error:", e)
```

## Study Case 8 - Baca File Nilai dan Hitung Rata-rata
File yang digunakan: `nilai_dummy.txt`.


```python
total = 0
jumlah = 0

try:
    with open("nilai_dummy.txt", "r", encoding="utf-8") as f:
        for baris in f:
            nama, nilai = baris.strip().split(",")
            nilai = int(nilai)
            print(f"{nama} -> {nilai}")
            total += nilai
            jumlah += 1

    rata_rata = total / jumlah
    print("Rata-rata nilai =", rata_rata)
except FileNotFoundError:
    print("File nilai_dummy.txt tidak ditemukan")
except ValueError:
    print("Format data nilai salah")
except Exception as e:
    print("Terjadi error:", e)
```

## Study Case 9 - Hidden Word Game (data langsung di kode)
Program memilih kata rahasia dari list, lalu user menebak huruf.


```python
kata_list = ["python", "regex", "library", "student", "kampus"]
kata_rahasia = random.choice(kata_list)

tebakan = ["_"] * len(kata_rahasia)
kesempatan = 6

print("Kata rahasia:", " ".join(tebakan))

while kesempatan > 0 and "_" in tebakan:
    huruf = input("Masukkan 1 huruf: ").lower()

    if len(huruf) != 1 or not huruf.isalpha():
        print("Input harus 1 huruf")
        continue

    if huruf in kata_rahasia:
        for i in range(len(kata_rahasia)):
            if kata_rahasia[i] == huruf:
                tebakan[i] = huruf
        print("Benar :", " ".join(tebakan))
    else:
        kesempatan -= 1
        print("Salah. Sisa kesempatan:", kesempatan)
        print("Kondisi:", " ".join(tebakan))

if "_" not in tebakan:
    print("Selamat, kamu menang. Kata:", kata_rahasia)
else:
    print("Game over. Kata rahasia:", kata_rahasia)
```

## Study Case 10 - Hidden Word Game dari File + Try Except
Kata diambil dari file `kata_rahasia.txt`.


```python
try:
    with open("kata_rahasia.txt", "r", encoding="utf-8") as f:
        daftar_kata = [baris.strip() for baris in f if baris.strip()]

    kata_rahasia = random.choice(daftar_kata)
    tebakan = ["_"] * len(kata_rahasia)
    kesempatan = 5

    print("Game dimulai")
    print("Kata:", " ".join(tebakan))

    while kesempatan > 0 and "_" in tebakan:
        huruf = input("Tebak huruf: ").lower()

        if len(huruf) != 1 or not huruf.isalpha():
            print("Masukkan hanya 1 huruf")
            continue

        if huruf in kata_rahasia:
            for i, ch in enumerate(kata_rahasia):
                if ch == huruf:
                    tebakan[i] = huruf
            print("Benar:", " ".join(tebakan))
        else:
            kesempatan -= 1
            print("Salah. Sisa:", kesempatan)

    if "_" not in tebakan:
        print("Menang! Kata rahasia =", kata_rahasia)
    else:
        print("Kalah. Kata rahasia =", kata_rahasia)

except FileNotFoundError:
    print("File kata_rahasia.txt tidak ditemukan")
except Exception as e:
    print("Terjadi error:", e)
```

```python
import re

data = {}

try:
    while True:
        nama = input("Nama panggilan: ").strip()
        email = input("Email: ").strip()

        if not nama:
            raise ValueError("Nama tidak boleh kosong")

        pattern = r'^[A-Za-z][A-Za-z0-9_]*@pemrograman\.id$'

        if not re.fullmatch(pattern, email):
            print("Email tidak valid")
        elif email in data.values():
            print("Email sudah terdaftar")
        else:
            data[nama] = email
            print(f"Data {nama} dengan email {email} berhasil ditambahkan")

        pilih = input("Tambah lagi? (y/n): ").strip().lower()
        if pilih == "n":
            break
        elif pilih != "y":
            raise ValueError("Pilihan hanya boleh y atau n")

    with open("pendaftaran_duplikat.txt", "w") as file:
        for nama, email in data.items():
            file.write(f"{nama}; {email}\n")

    print("\nData berhasil disimpan")

    with open("pendaftaran_duplikat.txt", "r") as file:
        print("\nIsi file:")
        print(file.read())

except ValueError as ve:
    print("ValueError:", ve)

except Exception as e:
    print("Error:", e)
```

```python
import re

# data awal yang sudah ada
data_kelompok = {
    "Kelompok A": [
        {"nama": "Andi", "email": "andi@pemrograman.id"},
        {"nama": "Budi", "email": "budi@pemrograman.id"}
    ],
    "Kelompok B": [
        {"nama": "Citra", "email": "citra@pemrograman.id"},
        {"nama": "Dina", "email": "dina@pemrograman.id"}
    ]
}

# set untuk menyimpan data email dan nama yang sudah ada
email_terdaftar = set()
nama_terdaftar = set()

# ambil data awal ke dalam set
for kelompok in data_kelompok:
    for anggota in data_kelompok[kelompok]:
        email_terdaftar.add(anggota["email"].lower())
        nama_terdaftar.add(anggota["nama"].lower())

try:
    jumlah_kelompok_baru = int(input("Masukkan jumlah kelompok baru yang ingin ditambahkan: "))

    if jumlah_kelompok_baru <= 0:
        raise ValueError("Jumlah kelompok harus lebih dari 0")

    # loop luar
    for i in range(jumlah_kelompok_baru):
        print(f"\n=== Input Kelompok Baru ke-{i+1} ===")
        nama_kelompok = input("Nama kelompok: ").strip()

        if not nama_kelompok:
            raise ValueError("Nama kelompok tidak boleh kosong")

        if nama_kelompok not in data_kelompok:
            data_kelompok[nama_kelompok] = []

        jumlah_anggota = int(input("Jumlah anggota yang ingin ditambahkan: "))

        if jumlah_anggota <= 0:
            raise ValueError("Jumlah anggota harus lebih dari 0")

        # loop dalam
        for j in range(jumlah_anggota):
            print(f"\nAnggota ke-{j+1}")
            nama = input("Nama: ").strip()
            email = input("Email: ").strip().lower()

            # regex nama: hanya huruf dan spasi
            pattern_nama = r'^[A-Za-z ]+$'
            # regex email: harus domain pemrograman.id
            pattern_email = r'^[a-zA-Z]+@pemrograman\.id$'

            if not re.fullmatch(pattern_nama, nama):
                print("Nama tidak valid, hanya boleh huruf dan spasi")
                continue

            if not re.fullmatch(pattern_email, email):
                print("Email tidak valid")
                continue

            if nama.lower() in nama_terdaftar:
                print("Nama sudah ada, tidak boleh duplikat")
                continue

            if email in email_terdaftar:
                print("Email sudah ada, tidak boleh duplikat")
                continue

            data_kelompok[nama_kelompok].append({
                "nama": nama,
                "email": email
            })

            nama_terdaftar.add(nama.lower())
            email_terdaftar.add(email)

            print(f"Data {nama} berhasil ditambahkan ke {nama_kelompok}")

    # tampilkan semua data
    print("\n=== HASIL AKHIR DATA KELOMPOK ===")
    for kelompok in data_kelompok:
        print(f"\n{kelompok}")
        for anggota in data_kelompok[kelompok]:
            print(f"- {anggota['nama']} ({anggota['email']})")

except ValueError as ve:
    print("ValueError:", ve)

except Exception as e:
    print("Terjadi error:", e)
```
