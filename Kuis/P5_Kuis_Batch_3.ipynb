```python
#pengecekan email duplikat

import re

def validasi_email(email):
    pattern_email = r'^[A-Za-z][A-Za-z0-9_]{1,}@pemrograman\.id$'
    return re.match(pattern_email, email)

data_peserta = []

while True:
    nama = input("Nama panggilan: ")
    email = input("Email: ")

    email_sudah_ada = False
    for data in data_peserta:
        if data[1] == email:
            email_sudah_ada = True
            break

    if not validasi_email(email):
        print("Email tidak valid")
    elif email_sudah_ada:
        print("Email sudah terdaftar")
    else:
        data_peserta.append((nama, email))
        print(f"Data {nama} berhasil ditambahkan")

    lagi = input("Tambah data lagi? (y/n): ").lower()
    if lagi != "y":
        break

print("\nDaftar peserta:")
for i, data in enumerate(data_peserta, start=1):
    print(f"{i}. {data[0]} - {data[1]}")
```

```python
#validasi_email pw

import re

def validasi_username(username):
    pattern = r'^[A-Za-z][A-Za-z0-9_]{4,}$'
    return re.match(pattern, username)

def validasi_password(password):
    pattern = r'^(?=.*[A-Z])(?=.*[a-z])(?=.*\d).{8,}$'
    return re.match(pattern, password)

username = input("Username: ")
password = input("Password: ")

if not validasi_username(username):
    print("Username tidak valid. Minimal 5 karakter, diawali huruf.")
elif not validasi_password(password):
    print("Password tidak valid. Minimal 8 karakter, ada huruf besar, huruf kecil, dan angka.")
else:
    print("Registrasi akun berhasil")
```

```python
#full program

import re

def validasi_email(email):
    pattern_email = r'^[A-Za-z][A-Za-z0-9_]{1,}@pemrograman\.id$'
    return re.match(pattern_email, email)

def tambah_data():
    nama = input("Nama: ")
    email = input("Email: ")

    if validasi_email(email):
        with open("../data/study/pendaftaran_menu.txt", "a") as f:
            f.write(f"{nama},{email}\n")
        print("Data berhasil ditambahkan")
    else:
        print("Email tidak valid")

def lihat_data():
    try:
        with open("../data/study/pendaftaran_menu.txt", "r") as f:
            print("\nData pendaftaran:")
            print(f.read())
    except FileNotFoundError:
        print("Belum ada data")

while True:
    print("\n=== MENU PENDAFTARAN ===")
    print("1. Tambah data")
    print("2. Lihat data")
    print("3. Keluar")

    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        tambah_data()
    elif pilihan == "2":
        lihat_data()
    elif pilihan == "3":
        print("Program selesai")
        break
    else:
        print("Pilihan tidak valid")
```
