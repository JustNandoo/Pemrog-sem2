```python
import re
import os

while True:
    files = []

    nickname = input("Nama panggilan: ")
    email = input("Email: ")

    if re.fullmatch(r'[a-zA-Z]+\@(pemrograman.id)', email) == None:
        print('Email tidak valid')

        repeat = input("Tambah data baru? (Y/N)")
        if repeat.upper() == "Y":
            continue
        else:
            break

    else:
        files.append(nickname + "; " + email)
        print(f'Data {nickname} dan {email} berhasil ditambahkan')

        repeat = input("Tambah data baru? (Y/N)")
        if repeat.upper() == "Y":
            continue
        else:
            break

save = open(os.getcwd(), 'w')
h = save.write(files)
save.close()
```

```python
import re

data = {}

try:
    while True:
        input_user = input("Masukkan nama panggilan dan email dalam satu baris dipisahkan spasi: ").strip()

        bagian = input_user.split()

        if len(bagian) < 2:
            raise IndexError("Input harus berisi nama dan email")

        nama_user = bagian[0]
        email_user = bagian[-1]

        pattern_email = r'^[a-zA-Z]+@pemrograman\.id$'
        match_email = re.search(pattern_email, email_user)

        if match_email:
            data[nama_user] = match_email.group()
            print(f"Data {nama_user} berhasil ditambahkan")
        else:
            print("Email invalid")

        lanjut = input("Tambah data baru? (y/n): ").strip().lower()

        if lanjut not in ["y", "n"]:
            raise ValueError("Input lanjut hanya boleh y atau n")

        if lanjut == "n":
            break

    with open("pendaftaran1.txt", "a") as file:
        for nama in data:
            file.write(f"{nama}; {data[nama]}\n")

    print("\nIsi file txt:")
    with open("pendaftaran1.txt", "r") as file:
        file_2 = file.read()
        print(file_2)

except IndexError as ie:
    print("Terjadi IndexError:", ie)

except ValueError as ve:
    print("Terjadi ValueError:", ve)

except FileNotFoundError:
    print("File tidak ditemukan")

except PermissionError:
    print("Tidak punya izin untuk membuka file")

except KeyboardInterrupt:
    print("\nProgram dihentikan oleh user")

except Exception as e:
    print("Terjadi kesalahan lain:", e)
```
