```python
import re
import os

while True:
    files = []

    nickname = input("Nama panggilan: ")
    email = input("Email: ")

    if re.fullmatch(r'[a-zA-Z]+\@(pemrograman.id)', email) == None:
        print('Email tidak valid')

        repeat = input("Tambah data baru? (Y/N)")
        if repeat.upper() == "Y":
            continue
        else:
            break

    else:
        files.append(nickname + "; " + email)
        print(f'Data {nickname} dan {email} berhasil ditambahkan')

        repeat = input("Tambah data baru? (Y/N)")
        if repeat.upper() == "Y":
            continue
        else:
            break

save = open(os.getcwd(), 'w')
h = save.write(files)
save.close()
```
