```python
import re

data = {}

while True:
    input_nama = input('Nama Panggilan: ')

    input_email = input('Email: ')

    pattern_email = r'^[a-zA-Z]+@pemrograman\.id$'

    match_email = re.search(pattern_email,input_email)

    if match_email:
        data[input_nama] = input_email 
    else:
        print('email tidak valid')

    lanjut = input('tambah data baru (y/n): ')
    if lanjut=='n':
        break

with open('pendaftaran.txt','a') as file:
    for nama in data:
        file.write(f'{nama}; {data[nama]}\n')

# menampilkan file
with open('pendaftaran.txt','r') as file:
    file_1 = file.read()

print(file_1)    
```

```python
# opsional 1, nama email satu baris
import re

data = {}

while True:
    input_user = input('Masukkan nama panggilan dan email dalam satu baris dipisahkan spasi: ').strip()

    nama_user = input_user.split(' ')[0]
    email_user = input_user.split(' ')[-1]

    pattern_email = r'\b[a-zA-Z]+@pemrograman\.id\b'
    match_email = re.search(pattern_email, email_user)

    if match_email:
        data[nama_user] = match_email[0] 
    else:
        print('email invalid')

    lanjut = input('tambah data baru (y/n): ')
    if lanjut=='n':
        break

with open('pendaftaran1.txt','a') as file:
    for nama in data:
        file.write(f'{nama}; {data[nama]}\n')

# menampilkan file
with open('pendaftaran1.txt','r') as file:
    file_2 = file.read()

print('isi file txt:\n' + file_2)
```

```python
# opsional 2, email gaboleh duplikat
import re

data = {}
data_email = []

while True:
    input_user = input('Masukkan nama panggilan dan email dalam satu baris dipisahkan spasi: ').strip()

    nama_user = input_user.split(' ')[0]
    email_user = input_user.split(' ')[-1]

    pattern_email = r'\b[a-zA-Z]+@pemrograman\.id\b'
    match_email = re.search(pattern_email, email_user)

    if email_user in data_email:        
        print('email sudah terdaftar, tidak disimpan')
    
    else:
        if match_email:
            data[nama_user] = match_email[0]
            data_email.append(email_user)

        else:
            print('email invalid, tidak disimpan')    

    lanjut = input('tambah data baru (y/n): ')
    if lanjut=='n':
        break

with open('pendaftaran2.txt','a') as file:
    for nama in data:
        file.write(f'{nama}; {data[nama]}\n')

# menampilkan file
with open('pendaftaran2.txt','r') as file:
    file_3 = file.read()

print('isi file txt:\n' + file_3)
```
