Pada kode ini, saya menggunakan modul re untuk melakukan pencarian pola teks dalam file. Kemudian saya membuka file berita.txt dan membaca seluruh isinya menggunakan read() agar seluruh teks dapat diproses. Setelah itu, saya membuat pola regex Prof\..+?([A-Z][a-z]+(?:\s[A-Z][a-z]+)+) untuk mengambil nama profesor, yaitu dimulai dari “Prof.” lalu diikuti nama yang huruf awalnya kapital dan terdiri dari minimal dua kata. Selanjutnya, saya menggunakan re.findall() untuk mengambil semua nama yang sesuai dengan pola tersebut. Hasil yang diperoleh kemudian dihilangkan duplikatnya menggunakan set() dan diurutkan secara alfabet menggunakan sorted(). Terakhir, hasil tersebut ditampilkan menggunakan print().

```python
import re

with open("../data/praktikum/berita.txt", "r") as file:
    text = file.read()

pattern = r'Prof\..+?([A-Z][a-z]+(?:\s[A-Z][a-z]+)+)'

names = re.findall(pattern, text)

sorted_names = sorted(set(names))

print(sorted_names)
```

Soal 5

Pada kode ini, saya menggunakan library pandas untuk mengolah data dan re untuk proses pembersihan teks. Pertama, saya membaca dataset reviews_dataset_updated.csv ke dalam dataframe df. Selanjutnya, saya membuat fungsi clean_text() yang berfungsi untuk menghapus semua karakter selain huruf dan spasi menggunakan regex [^a-zA-Z\s], sehingga tanda baca dan angka akan dihilangkan. Fungsi tersebut kemudian diterapkan pada kolom review_text menggunakan .apply() dan hasilnya disimpan ke kolom baru punctuation_and_digit_removal. Setelah itu, saya melakukan proses case folding dengan mengubah seluruh teks menjadi huruf kecil menggunakan .str.lower() dan menyimpannya ke kolom casefolding. Terakhir, dataframe yang sudah diproses disimpan ke file baru cleaned.csv tanpa menyertakan index, dan ditampilkan beberapa baris awalnya menggunakan print(df.head()).

```python
import pandas as pd
import re

df = pd.read_csv("../data/praktikum/reviews_dataset_updated.csv")

def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

df['punctuation_and_digit_removal'] = df['review_text'].apply(clean_text)

df['casefolding'] = df['punctuation_and_digit_removal'].str.lower()

df.to_csv("cleaned.csv", index=False)

print(df.head())
```
