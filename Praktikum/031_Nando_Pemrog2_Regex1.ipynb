## Soal 1

Di soal pertama ini saya membuat fungsi untuk mengecek apakah sebuah nickname valid atau tidak. Aturannya sederhana, nickname hanya boleh terdiri dari tepat 5 huruf kecil saja, tidak boleh ada angka, simbol, spasi, atau huruf kapital.

Pattern yang saya pake itu  adalah `^[a-z]{5}$`. Tanda `^` di awal dan `$` di akhir itu fungsinya untuk memastikan regex mengecek seluruh string dari awal sampai akhir, bukan hanya sebagian. Lalu `[a-z]` artinya hanya huruf kecil a sampai z yang diterima, dan `{5}` berarti harus tepat 5 karakter, tidak kurang tidak lebih.

Hasilnya udah sesuai. Nickname seperti `anaaa` atau `arumi` dianggap valid karena memenuhi semua syarat. Sedangkan `andragogi` invalid karena terlalu panjang, `anda` invalid karena kurang dari 5 huruf, `Arumi` invalid karena ada huruf kapital, dan `sania99` invalid karena ada angka di belakangnya.

```python
#SOAL 1

import re

def check_nickname(name):
    pattern = r'^[a-z]{5}$'

    if re.match(pattern, name):
        return "Valid"
    else:
        return "Invalid"

names = ["anaaa", "arumi", "sania", "vikar", "andragogi",
         "anda", "an_a", "Arumi", "sania99", "dian.a", "@diana"]

for n in names:
    print(n, ":", check_nickname(n))
```

## Soal 2

Soal kedua ini itu Format yang valid harus diawali dengan tepat 2 digit angka, lalu diikuti oleh satu atau lebih huruf (boleh huruf besar maupun kecil) atau titik. Tidak boleh ada underscore, angka di tengah atau di akhir, maupun simbol lainnya.

Pattern yang saya gunakan adalah `^[0-9]{2}[A-Za-z.]+$`. Bagian `[0-9]{2}` itu dia memastikan dua digit angka ada di awal. Lalu `[A-Za-z.]+` berarti setelah angka harus ada minimal satu karakter berupa huruf atau titik. Saya sengaja memasukkan titik ke dalam kurung kotak `[]` karena di dalam sana titik tidak perlu di-escape, dia langsung terbaca sebagai titik biasa.

Hasilnya udah benar. Nama seperti `11anaaa` atau `67dian.a` valid karena formatnya sesuai. Sedangkan `andara11` invalid karena angkanya ada di belakang bukan di depan, `9sania` invalid karena angkanya cuma satu digit, dan `10Aruumi10` itu bakal invalid karena ada angka lagi di akhir string.

```python
#SOAL 2

import re

def check_nickname(name):
    pattern = r'^[0-9]{2}[A-Za-z.]+$'

    if re.match(pattern, name):
        return "Valid"
    else:
        return "Invalid"

names = ["11anaaa", "18arum", "30Sania", "09vika", "43Andreas", "67dian.a",
         "andara11", "__andri", "Arumi10", "9sania", "dian.a", "10Aruumi10"]

for n in names:
    print(n, ":", check_nickname(n))
```

## Soal 3

Di soal ini saya membuat validasi khusus untuk email resmi PENS dengan domain `@pens.ac.id`. Aturannya, karakter pertama harus berupa huruf, lalu setelahnya boleh kombinasi huruf, angka, atau underscore dengan minimal satu karakter, dan domainnya harus tepat `@pens.ac.id`.

Pattern yang saya pakai adalah `^[A-Za-z][A-Za-z0-9_]{1,}@pens\.ac\.id$`. Bagian `[A-Za-z]` di awal memastikan karakter pertama adalah huruf. Lalu `[A-Za-z0-9_]{1,}` berarti minimal ada satu karakter lagi setelahnya. Yang penting di sini adalah penulisan `\.` untuk titik pada domain. Saya harus memakai backslash sebelum titik karena kalau tidak, titik di regex artinya karakter apapun, sehingga domain yang salah pun bisa lolos.

Hasilnya itu, Email seperti `Deby@pens.ac.id` dan `lely_aja@pens.ac.id` valid. Sementara `7dian@pens.ac.id,` invalid karena diawali angka dan ada koma di akhir, `d@pens.ac.id` invalid karena bagian lokalnya hanya satu karakter, dan `deby@pens.com` invalid karena domainnya berbeda.

```python
#SOAL 3

import re

def check_validity_email(email):
    pattern_email = r'^[A-Za-z][A-Za-z0-9_]{1,}@pens\.ac\.id$'

    if re.match(pattern_email, email):
        return "Valid"
    else:
        return "Invalid"

email = ["Deby@pens.ac.id", "amai78@pens.ac.id", "lely_aja@pens.ac.id", "d4@pens.ac.id", "7dian@pens.ac.id,", "d@pens.ac.id", "deby@pens.com", "_deby@pens "]

for e in email:
    print(e, ":", check_validity_email(e))
```

Saya menyiapkan dua pattern. `pattern_1` yaitu `\b(?:[A-Z]{2,})\b` digunakan untuk menangkap singkatan standar seperti `HTML`, `PENS`, atau `UI`. Tanda `\b` di kedua sisi fungsinya sebagai word boundary, artinya regex hanya akan mencocokkan kata utuh, bukan huruf kapital yang ada di tengah-tengah kata lain. Lalu `pattern_2` adalah versi yang lebih lengkap karena juga bisa menangkap singkatan berformat titik seperti `A.`, tapi untuk kode ini saya menggunakan `pattern_1`.

Setelah semua singkatan ditemukan menggunakan `re.findall()`, saya membungkusnya dengan `set()` lalu diubah kembali ke `list()`. Tujuannya supaya tidak ada duplikat, jadi kalau kata `HTML` muncul berkali-kali di teks, yang ditampilkan tetap satu saja.
