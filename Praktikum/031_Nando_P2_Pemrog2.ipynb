EXERCISE 1

- Pada Program Exercise 1 ini menggunakan 1 library, yaitu library datetime dengan mengimport class date agar bisa membuat tanggal
- mendefinisikan variabel year, month, day dalam bentuk tipe data integer
- setelah itu mengubah objek tanggal ( year, month, day ) menjadi string dengan format ISO
- Setelah itu tinggal di print saja dan akan menghasilkan output yang diiinginkan

```python
from datetime import date

year = 2021
month = 3
day = 25

d = date(year, month, day)

iso_date = d.isoformat()

print(iso_date)
```

EXERCISE 4

- Import class date dari library datetime untuk membuat dan mengelola tanggal

#A
- Memuat tanggal di date(2021, 3, 25)
- Menggunakan library date untuk memuat tanggal hari ini
- Menghitung hasil selisih hari dari variabel particular data dan variabel today
- difference.days itu berfungsi untuk menampilkan hasil selisih dalam bentuk hari

#B
- Mengambil nama hari dari tanggal particular date menggunakan format %A. %A itu adalah format untuk menampilkan nama hari dari suatu tanggal

```python
from datetime import date

#A
particular_date = date(2021, 3, 25)
today = date.today()
difference = today - particular_date

print("A :")
print("Today:", today)
print("Particular Date:", particular_date)
print("Difference :", difference.days, "days")


#B
day_name = particular_date.strftime("%A")

print("B :")
print("Date:", particular_date)
print("Day:", day_name)
```

EXERCISE 5

- Mengimport class date dari library datetime untuk menampilkan tanggal hari ini
- Mengimport library numpy yang dialiaskan "np" untuk perhitungan matematika
- variabel todaye itu untuk mengambil tanggal hari ini
- variabel numbers membuat array menggunakan numpy yang berisi data angka
- variabel average untuk menghitung rata rata dari data pakai fungsi mean
- variabel maximum untuk mencari nilai terbesar dari data pakai fungsi max
- variabel total untuk menghitung total jumlah angka dari data pakai fungsi sum


```python
from datetime import date
import numpy as np

today = date.today()

numbers = np.array([10, 20, 30, 40, 50])

average = np.mean(numbers)
maximum = np.max(numbers)
total = np.sum(numbers)

print(f"Pada tanggal {today}, program akan menghitung data berikut:")
print(f"Data angka: {numbers}")
print(f"Total jumlah data adalah {total}")
print(f"Rata-rata dari data tersebut adalah {average}")
print(f"Nilai maksimum dari data tersebut adalah {maximum}")
```
