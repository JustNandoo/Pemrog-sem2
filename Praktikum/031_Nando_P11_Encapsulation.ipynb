```python
class Restaurant:
    def __init__(self, name, cuisine_type):
        self.name = name
        self.cuisine_type = cuisine_type
        self.__menu = {}

    def describe_restaurant(self):
        print(f"Nama restoran: {self.__name}")
        print(f"Jenis masakan: {self.__cuisine_type}")

    def open_restaurant(self):
        print(f"Restoran {self.__name} telah dibuka.")

    def add_menu(self, menu_name, price):
        self.__menu[menu_name] = price

    def display_menu(self):
        print(f"\nDaftar Menu di {self.__name}:")
        if len(self.__menu) == 0:
            print("Belum ada menu yang tersedia.")
        else:
            for menu_name, price in self.__menu.items():
                print(f"- {menu_name}: Rp{price}")


restaurant1 = Restaurant("Ottoman Cafe", "International Cuisine")
restaurant2 = Restaurant("Dilan Resto", "Indonesian Cuisine")


restaurant1.add_menu("Nasi Goreng", 25000)
restaurant1.add_menu("Ayam Geprek", 22000)
restaurant1.add_menu("Es Teh Manis", 8000)


restaurant2.add_menu("Sushi", 45000)
restaurant2.add_menu("Ramen", 50000)
restaurant2.add_menu("Ocha", 10000)


restaurant1.describe_restaurant()
restaurant1.display_menu()

restaurant2.describe_restaurant()
restaurant2.display_menu()
```

```python
class Restaurant:
    def __init__(self, name, cuisine_type):
        self.name = name
        self.cuisine_type = cuisine_type
        self.__menu = {}   # atribut private untuk menyimpan menu dan harga

    def describe_restaurant(self):
        print(f"Nama restoran: {self.__name}")
        print(f"Jenis masakan: {self.__cuisine_type}")

    def open_restaurant(self):
        print(f"Restoran {self.__name} telah dibuka.")

    def add_menu(self, menu_name, price):
        self.__menu[menu_name] = price
        print(f"Menu {menu_name} berhasil ditambahkan.")

    def remove_menu(self, menu_name):
        if menu_name in self.__menu:
            del self.__menu[menu_name]
            print(f"Menu {menu_name} berhasil dihapus.")
        else:
            print(f"Menu {menu_name} tidak ditemukan.")

    def display_menu(self):
        print(f"\nDaftar menu di {self.__name}:")
        if len(self.__menu) == 0:
            print("Belum ada menu yang tersedia.")
        else:
            for menu_name, price in self.__menu.items():
                print(f"- {menu_name}: Rp{price}")


restaurant1 = Restaurant("Ottoman Cafe", "Indonesian Food")
restaurant2 = Restaurant("Golden Lamian Ramen", "Japanese Food")


restaurant1.add_menu("Nasi Goreng", 25000)
restaurant1.add_menu("Ayam Geprek", 22000)
restaurant1.add_menu("Es Teh Manis", 8000)

restaurant2.add_menu("Sushi", 45000)
restaurant2.add_menu("Ramen", 50000)
restaurant2.add_menu("Ocha", 10000)


restaurant1.display_menu()
restaurant2.display_menu()


restaurant1.remove_menu("Ayam Geprek")
restaurant2.remove_menu("Ocha")


restaurant1.display_menu()
restaurant2.display_menu()
```

```python
class User:
    def __init__(self, first_name, last_name, age, email, username):
        self.__first_name = first_name
        self.__last_name = last_name
        self.__age = age
        self.__email = email
        self.__username = username
        self.__login_attempt = 0   # nilai default login attempt

    def describe_user(self):
        print("Informasi User:")
        print(f"Nama lengkap  : {self.__first_name} {self.__last_name}")
        print(f"Umur          : {self.__age}")
        print(f"Email         : {self.__email}")
        print(f"Username      : {self.__username}")
        print(f"Login Attempt : {self.__login_attempt}")
        print()

    def greet_user(self):
        print(f"Halo, {self.__first_name} {self.__last_name}! Selamat datang kembali.")
        print()

    def increment_login_attempt(self):
        self.__login_attempt += 1

    def reset_login_attempt(self):
        self.__login_attempt = 0

    def display_login_attempt(self):
        print(f"Jumlah login attempt: {self.__login_attempt}")


# Membuat objek dari class User
user1 = User("Abdullah", "Wijayanto", 25, "Abdul@gmail.com", "abdulbolos")


# Menampilkan data user
user1.describe_user()
user1.greet_user()


# Uji coba method increment_login_attempt()
user1.increment_login_attempt()
user1.increment_login_attempt()
user1.increment_login_attempt()

print("Login Attempt:")
user1.display_login_attempt()


# Uji coba method reset_login_attempt()
user1.reset_login_attempt()

print("Setelah login attempt di-reset:")
user1.display_login_attempt()
```

```python
class User:
    def __init__(self, first_name, last_name, age, email, username):
        self.__first_name = first_name
        self.__last_name = last_name
        self.__age = age
        self.__email = email
        self.__username = username
        self.__login_attempt = 0   # atribut private

    def describe_user(self):
        print("Informasi User:")
        print(f"Nama lengkap  : {self.__first_name} {self.__last_name}")
        print(f"Umur          : {self.__age}")
        print(f"Email         : {self.__email}")
        print(f"Username      : {self.__username}")
        print(f"Login Attempt : {self.__login_attempt}")
        print()

    def greet_user(self):
        print(f"Halo, {self.__first_name} {self.__last_name}! Selamat datang kembali.")
        print()

    def increment_login_attempt(self):
        self.__login_attempt += 1

    def reset_login_attempt(self):
        self.__login_attempt = 0

    def display_login_attempt(self):
        print(f"Jumlah login attempt: {self.__login_attempt}")


# Membuat objek user
user1 = User("Joni", "Benedictus", 40, "Joni@gmail.com", "waleleng")


# Menampilkan data awal
user1.describe_user()


# Menambah login attempt melalui method
user1.increment_login_attempt()
user1.increment_login_attempt()
user1.increment_login_attempt()

print("Setelah increment_login_attempt() dipanggil 3 kali:")
user1.display_login_attempt()


# Mencoba mengubah login_attempt dari luar class
user1.__login_attempt = 100

print("\nSetelah mencoba mengubah __login_attempt dari luar class:")
user1.display_login_attempt()


# Reset login attempt melalui method
user1.reset_login_attempt()

print("\nSetelah reset_login_attempt() dipanggil:")
user1.display_login_attempt()
```

```python
class User:
    __max_login_attempt = 3

    def __init__(self, first_name, last_name, age, email, username):
        self.__first_name = first_name
        self.__last_name = last_name
        self.__age = age
        self.__email = email
        self.__username = username
        self.__login_attempt = 0

    def describe_user(self):
        print("Informasi User:")
        print(f"Nama lengkap       : {self.__first_name} {self.__last_name}")
        print(f"Umur               : {self.__age}")
        print(f"Email              : {self.__email}")
        print(f"Username           : {self.__username}")
        print(f"Login Attempt      : {self.__login_attempt}")
        print(f"Max Login Attempt  : {User.__max_login_attempt}")
        print()

    def greet_user(self):
        print(f"Halo, {self.__first_name} {self.__last_name}! Selamat datang kembali.")
        print()

    def increment_login_attempt(self):
        self.__login_attempt += 1

        if self.__login_attempt > User.__max_login_attempt:
            print(f"Notifikasi: {self.__username} sudah melebihi batas maksimal login attempt!")

    def reset_login_attempt(self):
        self.__login_attempt = 0

    def display_login_attempt(self):
        print(f"Login attempt {self.__username}: {self.__login_attempt}")

    @classmethod
    def set_max_login_attempt(cls, max_attempt):
        cls.__max_login_attempt = max_attempt
        print(f"Max login attempt berhasil diubah menjadi {cls.__max_login_attempt}")

    @classmethod
    def display_max_login_attempt(cls):
        print(f"Max login attempt saat ini: {cls.__max_login_attempt}")


# Membuat 2 objek dari class User
user1 = User("Paul", "Klein", 25, "LANY@gmail.com", "Alonica")
user2 = User("Abel", "Tesfaye", 31, "weeknd@gmail.com", "Stargirl")


# Menampilkan data awal
user1.describe_user()
user2.describe_user()


# Uji coba login attempt user1
user1.increment_login_attempt()
user1.increment_login_attempt()
user1.increment_login_attempt()
user1.increment_login_attempt()  # login ke-4, melebihi batas maksimal

user1.display_login_attempt()


# Menampilkan max login attempt
User.display_max_login_attempt()


# Mengubah nilai max_login_attempt untuk semua objek
User.set_max_login_attempt(5)


# Setelah max_login_attempt diubah, efeknya berlaku untuk semua objek
user1.describe_user()
user2.describe_user()
```
