```python
hidden_word = input("Hidden word: ")
word = input("Word: ")

def find_hidden_word(hidden_word, word):
    try:
        if not hidden_word or not word:
            raise Exception("Error: input can't be empty")

        hidden_word = hidden_word.lower()
        word = word.lower()

        index = 0

        for char in hidden_word:
            found_at = word.find(char, index)

            if found_at == -1:
                raise Exception(f"Not found")

            index = found_at + 1

        print("Found")

    except Exception as e:
        print(e)

find_hidden_word(hidden_word, word)
```

```python
def read_accounts(filename):
    accounts = {}
    try:
        with open(filename, "r") as file:
            for line in file:
                line = line.strip()
                if line:
                    acc_number, name, pin, balance = line.split("|")
                    accounts[acc_number] = {
                        "name": name,
                        "pin": pin,
                        "balance": int(balance)
                    }
        return accounts
    except FileNotFoundError:
        print("Error: File Not Found")
        return None


def save_accounts(filename, accounts):
    with open(filename, "w") as file:
        for acc_number, data in accounts.items():
            file.write(f"{acc_number}|{data['name']}|{data['pin']}|{data['balance']}\n")


def login(accounts):
    acc_number = input("Enter account number: ")

    if acc_number not in accounts:
        raise Exception("Error: Account Not Found")

    pin = input("Enter PIN (6 digits): ")

    if not pin.isdigit() or len(pin) != 6:
        raise Exception("Error: Invalid PIN Format")

    if pin != accounts[acc_number]["pin"]:
        raise Exception("Error: Incorrect PIN")

    return acc_number


def transaction_menu():
    print("\nTransaction Menu")
    print("1. Deposit")
    print("2. Withdraw")
    choice = input("Choose transaction (1/2): ")
    return choice


def process_transaction(accounts, acc_number):
    choice = transaction_menu()

    try:
        amount = int(input("Enter amount: "))
        if amount <= 0:
            raise Exception("Error: Invalid Transaction Amount")
    except ValueError:
        raise Exception("Error: Invalid Transaction Amount")

    if choice == "1":
        accounts[acc_number]["balance"] += amount
        print("Deposit successful")
        print("New balance:", accounts[acc_number]["balance"])

    elif choice == "2":
        if amount > accounts[acc_number]["balance"]:
            raise Exception("Error: Invalid Transaction Amount")
        accounts[acc_number]["balance"] -= amount
        print("Withdrawal successful")
        print("New balance:", accounts[acc_number]["balance"])

    else:
        raise Exception("Error: Invalid Menu Choice")


filename = "../data/praktikum/accounts.txt"
accounts = read_accounts(filename)

if accounts is not None:
    try:
        acc_number = login(accounts)
        print(f"\nWelcome, {accounts[acc_number]['name']}")
        print("Current balance:", accounts[acc_number]["balance"])
        process_transaction(accounts, acc_number)
        save_accounts(filename, accounts)
    except Exception as e:
        print(e)
```

```python
import math


def calculator():
    while True:
        try:
            print("\n=== ADVANCED CALCULATOR ===")
            print("1. Addition (+)")
            print("2. Subtraction (-)")
            print("3. Multiplication (*)")
            print("4. Division (/)")
            print("5. Square Root (√)")
            print("6. Exit")

            choice = input("Choose menu: ")

            if choice == "6":
                print("Program ended")
                break

            if choice not in ["1", "2", "3", "4", "5"]:
                raise Exception("Error: Invalid menu choice")

            # operasi akar hanya butuh 1 angka
            if choice == "5":
                num = float(input("Enter number: "))
                if num < 0:
                    raise ValueError("Error: Cannot sqrt negative number")
                print("Result:", math.sqrt(num))

            else:
                a = float(input("Enter first number: "))
                b = float(input("Enter second number: "))

                if choice == "1":
                    print("Result:", a + b)
                elif choice == "2":
                    print("Result:", a - b)
                elif choice == "3":
                    print("Result:", a * b)
                elif choice == "4":
                    if b == 0:
                        raise ZeroDivisionError("Error: Division by zero")
                    print("Result:", a / b)

        except ValueError:
            print("Error: Invalid number input")

        except ZeroDivisionError:
            print("Error: Cannot divide by zero")

        except Exception as e:
            print(e)


calculator()
```
