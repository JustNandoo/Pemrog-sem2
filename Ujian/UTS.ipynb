```python
from re import findall, fullmatch

data = {}

def load_file():
    with open("../data/UTS/research_activities.txt", "r") as file:
        text = file.read()

    pattern = r'(\d{2}-\d{2}-\d{4}).*?([A-Z]{2,})'

    results = findall(pattern, text)

    for full_date, pt in results:
        date_key = full_date[:5]

        if date_key not in data:
            data[date_key] = []

        if pt not in data[date_key]:
            data[date_key].append(pt)


load_file()

while True:
    try:
        date = input("Tanggal aktivitas (dd-mm) : ")

        if not fullmatch(r'\d{2}-\d{2}', date):
            raise ValueError('Format penulisan tanggal salah')

        if date not in data:
            raise KeyError(date)

        print(f'Daftar kampus yang terlibat pada tanggal {date} adalah {data[date]}')

    except ValueError as e:
        print(f'Format penulisan tanggal salah')

    except KeyError:
        print(f'Tidak ada data untuk tanggal tersebut')

    if input("Ulangi lagi (y/t) ? ").lower() != 'y':
        break
```
